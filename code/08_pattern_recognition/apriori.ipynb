```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('../01_preprocess/files_ps/ycsb_c_ps.csv')

# Display the DataFrame
df
```

```python
df['percentile bracket'] = pd.qcut(df['time'], q=10, labels=False, duplicates='drop') + 1
df.drop([ 'time', 'rand_seed'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=df.columns)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.3, use_colnames=True, low_memory=True)
```
