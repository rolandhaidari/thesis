```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('../01_preprocess/files_just_ps/ycsb_c_just_ps.csv')

# Display the DataFrame
df
```

```python
df = df[df['const_enableHash'] != 1]
df = df[df['const_enableDense'] != 1]
df
```

```python
df['percentile bracket'] = pd.qcut(df['time'], q=3, labels=False, duplicates='drop') + 1
df.drop([ 'time', 'rand_seed'], axis=1, inplace=True)

columns_to_drop = df.columns[df.nunique() == 1]
df = df.drop(columns=columns_to_drop)
```

```python
# Identify columns with only True and False
boolean_columns = df.columns[(df.eq(True) | df.eq(False)).any()]

# Identify columns without only True and False
non_boolean_columns = df.columns.difference(boolean_columns)
print(non_boolean_columns)
# Apply one-hot encoding only to non-boolean columns
df = pd.get_dummies(df, columns=non_boolean_columns)

df = pd.get_dummies(df, columns=['density'])

# Convert True to 1 and False to 0 in boolean columns
df = df.astype(int)
```

```python
df = pd.get_dummies(df, columns=['percentile bracket'])
df
```

```python
#df = df[0:100]
df=df.astype(int)
df
```

```python
df.columns
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets =apriori(df, min_support=0.10, use_colnames=True, low_memory=True, verbose=1)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

frequent_itemsets
```

```python

pd.options.display.max_colwidth = 200
filtered_df = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: any('percentile' in item for item in x))]
filtered_df
```

```python
from mlxtend.frequent_patterns import association_rules
ar=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
ar
```

```python
o = ar[ar['consequents'].apply(lambda x: any('percentile' in item for item in x))]
o
```

```python
o_sorted = o.sort_values(by='confidence', ascending=False)
o_sorted
```
